```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
with open('abc.json', 'r') as f:
    data=f.read()

data=json.loads(data)
type(data)
```

```python
data.keys()
```

```python
df1=pd.DataFrame(data['surveyMeta']).T
df1.head(2)
```

```python
for col in df1.columns:
    print(col, type(df1[col][0]))
```

```python
df2= pd.DataFrame.from_dict(data['submissions'])
df2.head(2)
```

```python
for col in df2.columns:
    print(col, type(df2[col][0]))
```

```python
df_list=[]
for i in range(len(df2['questionValues'])):
    test=pd.DataFrame(df2['questionValues'][i]).T
    df_list.append(test)
    
df3=pd.concat(df_list, axis=1, sort=False)
df3.head(2)
```

```python
df4=df3['response']
df4.head(2)
```

```python
df5=df1.merge(df4,how='inner',left_on=df1.index,right_on=df4.index)
df5.head(2)
```

```python
df6=df5[['id','question','response']]
df6.columns=['id', 'question']+ ['response_'+str(i) for i in range(1,len(df6.columns[2:])+1)]
df6.head(2)
```

```python
type_lists=[]
for i in range(df6.shape[0]):
    row=df6.loc[i]
    if any(row.apply(lambda x:type(x)).values==list):
        type_lists.append(row['id'])
type_lists
```

```python
df6.set_index('id',inplace=True)
```

```python
df7=df6.loc[type_lists]
df7
```

```python
def get_value(a):
    if len(a)>1:
        return a[1]
    else:
        return ''
    
resp=[]    
for col in df7.columns[1:]:
    temp=[]
    for i in range(len(df7[col])):
        li=df7[col].values[i]
        for each in li:
            temp.append(get_value(each))
    resp.append(temp)
    
resp_df=pd.DataFrame(resp).T    
resp_df.columns=df7.columns[1:]


field=[]
for i in range(len(df7['response_1'].values)):
    for j in range(len(df7['response_1'].values[i])):
        field.append(df7['response_1'].values[i][j][0])
        
resp_df['field']=field

ids=[]
ques=[]
for i,each in enumerate(df7['response_1'].apply(lambda x: len(x))):
    for c in range(each):
        ids.append(df7['response_1'].apply(lambda x: len(x)).index[i])
        ques.append(df7['question'].values[i])
        
resp_df['id']=ids
resp_df['question']=ques

resp_df.set_index('id',inplace=True)
resp_df.head(2)
```

```python
df8=df6.loc[set(df6.index)-(set(type_lists))]
df8['field']=None

result=pd.concat([df8,resp_df],sort=False)
result.reset_index(inplace=True)
result.set_index(['id','question','field'])
```

```python
result.to_csv('result.csv')
```
